Написать линейную регрессию на два атрибута. Допустим, мы хотим предсказать цену дома (MEDV) в Бостоне, зная что цена на дом имеет линейную зависимость с атрибутом LSTAT. 

```python
medv = np.array(data.MEDV).reshape(-1,1)
lstat = np.array(data.LSTAT).reshape(-1,1)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```



1.   Разделить данные на 80% тренировочных, 20% тестовых
2.   Составить модель линейной регрессии
3.   Посчитать ошибку модели на тестовых данных
4.   Сделать точечный график данных и полученной прямой линейной регрессии



```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    medv,
    lstat,
    test_size = 0.2
)
```

```python
model.fit(X_train, y_train)
```

```python
predicted_y = model.predict(X_test)
```

```python
print('Средняя квадратичная ошибка: ', mean_squared_error(predicted_y, y_test))
print('Средняя абсолютная ошибка: ',mean_absolute_error(predicted_y, y_test))
```

```python
plt.xlabel('MEDV')
plt.ylabel('LSTAT')
plt.title('Линейная регрессия от одной переменной')
# Построим прямую для регрессии
# Создадим 100 точек для предсказаний
x = np.linspace(-50, 200, 100)
# Посчитаем предсказанные значения
# y = kx + b
# k = model.coef_
# b = model.intercept_
y = model.coef_*x + model.intercept_
y = np.transpose(y)
plt.plot(x,y)
plt.plot(X_train, y_train, 'r.')
plt.plot(X_test, y_test, 'go')
plt.xlim(0,60)
plt.ylim(-10,60)
plt.legend(['Прямая линейной регрессии', 'Тренировочные данные', 'Тестовые данные'])
plt.show()
```

# Дискретное распределение случайной величины

Напишите код для моделирования 10 подбрасываний игрального кубика. Начертите диаграму histogram. Какое это распределение?

```python
import pylab
import numpy

cast = 10
values = numpy.random.randint(1, 7, size=cast)
pylab.hist(values, bins=[0.5, 1.5, 2.5, 3.5, 4.5, 5.5, 6.5])

pylab.xlabel('Выпавшее число')
pylab.ylabel('Количество бросков')
pylab.title('Подбрасывание кубика 10 раз') 
pylab.show()
```

Диаграмма подбрасований кубика 10 раз имеет многовершинное распределение

Смоделируйте 1000 раз сумму по 100 подбрасываний. Начертите диаграму histogram. Почему график изменился? Почему можно применить центральную предельную теорему?

```python
import pylab
import numpy

cast = 100
values = numpy.random.randint(1, 7, size=cast)
for i in range (1000):
    values = values + numpy.random.randint(1, 7, size=cast)
pylab.hist(values/1000, bins=[0.5, 1.5, 2.5, 3.5, 4.5, 5.5, 6.5])

pylab.xlabel('Выпавшее число')
pylab.ylabel('Количество бросков')
pylab.title('Подбрасывание кубика 100 раз') 
pylab.show()
```

График поменялся так как сумма рандомных независимых величин стримится к нормальному распределению. А так как данный график 
стремится к нормальному распределению, что соответсвует утверждению централизованной предельной теореме, возможно ее применение.
