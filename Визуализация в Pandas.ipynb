```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('RU_youtube_trending_data.csv')
df.head(60)
```

```python
sns.jointplot(x='view_count', y='comment_count', 
              data=df, kind='reg')
```
